```python
import matplotlib.pyplot as plt

years = [2018, 2019, 2020, 2021, 2022]
noi = [200, 220, 250, 270, 300]     # Khoa Ni
ngoai = [150, 160, 180, 210, 230]   # Khoa Ngoi
nhi = [100, 120, 150, 180, 190]     # Khoa Nhi

# --- Biu  line chart ---
plt.figure(figsize=(10,5))
plt.plot(years, noi, label="Khoa Ni")
plt.plot(years, ngoai, label="Khoa Ngoi")
plt.plot(years, nhi, label="Khoa Nhi")
plt.title("Line Chart  S ca bnh theo tng khoa")
plt.xlabel("Nm")
plt.ylabel("S ca bnh")
plt.xticks(years)   # p ch hin nm nguyn
plt.legend()
plt.show()

# --- Biu  stackplot ---
plt.figure(figsize=(10,5))
plt.stackplot(years, noi, ngoai, nhi, labels=["Khoa Ni", "Khoa Ngoi", "Khoa Nhi"])
plt.title("Stackplot  S ca bnh tch ly theo khoa")
plt.xlabel("Nm")
plt.ylabel("Tng s ca bnh")
plt.xticks(years)   # p ch hin nm nguyn
plt.legend(loc="upper left")
plt.show()
```

```python
import matplotlib.pyplot as plt

# D liu gi lp huyt p (mmHg)
noi = [120, 125, 130, 140, 135, 150, 145, 155, 160, 170,200,60]
ngoai = [110, 115, 120, 125, 128, 130, 135, 140, 145, 150]
nhi = [90, 95, 100, 105, 98, 102, 110, 115, 120, 125]

data = [noi, ngoai, nhi]

plt.boxplot(data, labels=["Ni", "Ngoi", "Nhi"])
plt.title("Phn b huyt p tm thu theo khoa")
plt.ylabel("Huyt p (mmHg)")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Gi tr trung bnh huyt p sau iu tr (3 nhm)
nhom = ["Thuc A", "Thuc B", "Gi dc"]
mean = [130, 125, 135]       # Trung bnh
error = [5, 7, 6]            # Sai s ( mmHg)

plt.errorbar(nhom, mean, yerr=error, fmt='o', capsize=5, color='blue')

plt.title("Kt qu th nghim lm sng (Huyt p trung bnh  sai s)")
plt.ylabel("Huyt p (mmHg)")
plt.show()
```

```python
import numpy as np

x = np.linspace(0,10,100)      # To 100 im t 0  10
y = np.sin(x)                  # Hm sin(x)

plt.plot(x, y)                 # V ng cong sin
plt.fill_between(x, y-0.1, y+0.1, alpha=0.3)  # T vng gia (y-0.1, y+0.1)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D   # Import cng c v 3D trong Matplotlib

X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)

X, Y = np.meshgrid(X, Y)     # To li ta   t X v Y

Z = np.sin(np.sqrt(X**2 + Y**2))
# Cng thc: Z = sin(r) vi r = (X + Y)
# -> Hm sng hnh trn ng tm lan ta t gc (0,0)

fig = plt.figure()   # To mt Figure

ax = fig.add_subplot(111, projection='3d')
#add_subplot() thm mt trc (Axes) vo trong Figure.
#1 hng,
#1 ct,
#1 v tr (tc l v 1 biu ).

ax.plot_surface(X, Y, Z, cmap='viridis')
# V b mt 3D (surface plot) vi ta  X, Y, Z
# cmap='viridis' = dng colormap Viridis

plt.show()
```
